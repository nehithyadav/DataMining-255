```python
import os
import pandas as pd
import numpy as np

from sodapy import Socrata
```

```python
domain = 'data.cdc.gov'
client=Socrata(domain, None)
```

```python
#https://data.cdc.gov/resource/x9gk-5huc.json

# Enter the information from those sections here NNDSS 2023

weekly_dataset_identifier = 'x9gk-5huc'
weekly_results = client.get(weekly_dataset_identifier)
weekly_df = pd.DataFrame.from_dict(weekly_results)
weekly_df.head()
```

```python
#https://data.cdc.gov/resource/45b4-9j7u.json

#infrequently reported notifiable diseases 2017

disease1_dataset_identifier = '45b4-9j7u'
disease1_results = client.get(disease1_dataset_identifier)
disease1_df = pd.DataFrame.from_dict(disease1_results)
disease1_df.head()
```

```python
#https://data.cdc.gov/resource/pb4z-432k.json
#2015 nndss

disease2_dataset_identifier = 'pb4z-432k'
disease2_results = client.get(disease2_dataset_identifier)
disease2_df = pd.DataFrame.from_dict(disease2_results)
disease2_df.head()
```

```python
#https://data.cdc.gov/resource/wcwi-x3uk.json
#nndss 2014
disease3_dataset_identifier = 'wcwi-x3uk'
disease3_results = client.get(disease3_dataset_identifier)
disease3_df = pd.DataFrame.from_dict(disease3_results)
disease3_df.head()
```

```python
#https://data.cdc.gov/resource/dwqk-w36f.json
#NNDSS 2016
disease4_dataset_identifier = 'dwqk-w36f'
disease4_results = client.get(disease4_dataset_identifier)
disease4_df = pd.DataFrame.from_dict(disease4_results)
disease4_df.head()
```

```python
# Data Cleaning
dfs = [disease3_df, disease2_df, disease4_df, disease1_df]

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.reset_index(drop=True, inplace=True)

merged_df.head()
```

```python
merged_df.to_gbq('project_1.merged_data', project_id='dmproject-404916', if_exists='replace')
```

```python
check_df=merged_df[['disease','mmwr_year','mmwr_week','total_cases_reported_2011','total_cases_reported_2009','total_cases_reported_2016','total_cases_reported_2015','total_cases_reported_2014','total_cases_reported_2013','total_cases_reported_2012','total_cases_reported_2010','_5_year_weekly_average']]
```

```python
check_df.head()
```

```python
import pandas as pd

check_df['_5_year_weekly_average'] = pd.to_numeric(check_df['_5_year_weekly_average'], errors='coerce')
```

```python
mean_value = check_df['_5_year_weekly_average'].mean()
```

```python
check_df['_5_year_weekly_average'].fillna(mean_value, inplace=True)

check_df['_5_year_weekly_average'] = check_df['_5_year_weekly_average'].astype(int)
```
